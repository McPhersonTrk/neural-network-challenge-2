```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
# create a function to encode the object columns
def encode_columns(X_df):
    encoder = OneHotEncoder(sparse_output=False)
    obj_cols = X_df.select_dtypes(include=['object']).columns.tolist()
    if obj_cols:  # Check if there are any columns with 'object' .dtypes
        encoded_data = encoder.fit_transform(X_df[obj_cols]) #OHE encode 'object' .dtype columns
        encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out()) # Convert to DataFrame
        encoded_df.index = X_df.index  # set index
        X_df = pd.concat([X_df.drop(columns=obj_cols), encoded_df], axis=1) # concat both df's and drop original columns
    return X_df
# call the function
X_df_encoded = encode_columns(X_df)
X_df_encoded
```

```python
from sklearn.preprocessing import StandardScaler
# train test split
X_train,X_test,y_train,y_test=train_test_split(X_df_encoded,y_df,random_state=42)

# Create a StandardScaler
StandardScaler=StandardScaler()
# Fit the StandardScaler to the training data
scaled_X_df=StandardScaler.fit(X_df_encoded)
# Scale the training and testing data
X_train_scaled=StandardScaler.transform(X_train)
X_test_scaled=StandardScaler.transform(X_test)
X_train_scaled
#X_test_scaled

```

```python
# Create a dataframe with the scaled data
X_train_scaled_df=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled_df=pd.DataFrame(X_test_scaled,columns=X_test.columns)
X_train_scaled_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
# Create a OneHotEncoder for the Department column
yD_encoder = OneHotEncoder(sparse_output=False)
# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data
yD_encoded_train= yD_encoder.fit_transform(y_train[['Department']])
yD_encoded_test= yD_encoder.fit_transform(y_test[['Department']])
yD_encoded_train
```

```python
# create a dataframe with the encoded "Department" data
yD_encoded_train_df=pd.DataFrame(yD_encoded_train,columns=yD_encoder.get_feature_names_out())
yD_encoded_test_df=pd.DataFrame(yD_encoded_test,columns=yD_encoder.get_feature_names_out())
yD_encoded_train_df.head()
```

```python
# Create a OneHotEncoder for the Attrition column
from sklearn.preprocessing import OneHotEncoder
# Create a OneHotEncoder for the Department column
yA_encoder = OneHotEncoder(sparse_output=False)
# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data
yA_encoded_train= yA_encoder.fit_transform(y_train[['Attrition']])
yA_encoded_test= yA_encoder.fit_transform(y_test[['Attrition']])
yA_encoded_train
```

```python
# create a dataframe with the encoded "Attrition" data
yA_encoded_train_df=pd.DataFrame(yA_encoded_train,columns=yA_encoder.get_feature_names_out())
yA_encoded_test_df=pd.DataFrame(yA_encoded_test,columns=yA_encoder.get_feature_names_out())
yA_encoded_train_df.head()
```

## Create, Compile, and Train the Model

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense

# Find the number of columns in the X training data
X_shape= X_train_scaled_df.shape[1] # the number of columns in the X training data is 33, & 1102 rows.
# Create the input layer
Input_shape= (33,)
input_layer = layers.Input(shape=(X_shape,), name='input_features')

# Create at least two shared layers
shared_layer1 = layers.Dense(33, activation='relu')(input_layer)
shared_layer2 = layers.Dense(33, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department with a hidden layer and an output layer.
Department_dense= layers.Dense(3, activation='relu')(shared_layer1)
# Create the hidden layer
Department_hidden = layers.Dense(64, activation='relu')(shared_layer2)
# Create the output layer
Department_output= layers.Dense(len(yD_encoded_train_df.columns),
                                activation='sigmoid', 
                                name='Department_output')(Department_dense)
```

```python
# Create a branch for Attrition with a hidden layer and an output layer.
Attrition_dense= layers.Dense(3, activation='relu')(shared_layer1)
# Create the hidden layer
Attrition_hidden = layers.Dense(64, activation='relu')(shared_layer2)
# Create the output layer
Attrition_output= layers.Dense(len(yA_encoded_train_df.columns),
                                activation='sigmoid', 
                                name='Attrition_output')(Attrition_dense)
```

```python
# Create the model
model= Model(inputs=input_layer, outputs=[Department_output, Attrition_output])
# Compile the model
model.compile(optimizer='adam',
                loss={'Department_output': 'categorical_crossentropy', 
                    'Attrition_output': 'binary_crossentropy'},
                metrics={
                    'Department_output': ['accuracy'],
                    'Attrition_output': ['accuracy']
                })

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled_df,
    {
        'Department_output': yD_encoded_train_df,
        'Attrition_output': yA_encoded_train_df.iloc[:, :2]  # Select only the first 2 columns of yA_encoded_train_df
    },
    epochs=100,
    batch_size=100,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
results= model.evaluate(X_test_scaled_df, {
                            'Department_output': yD_encoded_test_df,
                            'Attrition_output': yA_encoded_test_df 
                        })
# Print the accuracy for both department and attrition
# Print the Quality and Description accuracy for the testing data
print(f"Department Accuracy: {results[3]:.4f}, Department Loss: {results[2]:.4f}")
print(f"Attrition Accuracy: {results[4]:.4f}, Attrition Loss: {results[2]:.4f}")
```
